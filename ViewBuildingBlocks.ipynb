```python
# Import the required libraries
import torch
import torchvision
from torchvision.io import read_image
from torchvision.utils import draw_bounding_boxes
from PIL import Image, ImageFont, ImageDraw 
from tkinter import *
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont

```

```python

# read input image
img = read_image('/Users/kashvi/Downloads/1907/Images/pano_0000_000059.jpg')

# bounding box in (xmin, ymin, xmax, ymax) format
# top-left point=(xmin, ymin), bottom-right point = (xmax, ymax)

bbox = [5718,3358,6035,4134]
bbox = torch.tensor(bbox, dtype=torch.int)
print(bbox)
print(bbox.size())
bbox = bbox.unsqueeze(0)
print(bbox.size())

bbox1= [2520,4515,3278,4633]
bbox1 = torch.tensor(bbox1, dtype=torch.int)
print(bbox1)
print(bbox1.size())
bbox1 = bbox1.unsqueeze(0)
print(bbox1.size())

bbox2= [4256,4134,5428,4279]
bbox2 = torch.tensor(bbox2, dtype=torch.int)
print(bbox2)
print(bbox2.size())
bbox2 = bbox2.unsqueeze(0)
print(bbox2.size())

# draw bounding box on the input image
img=draw_bounding_boxes(img, bbox, width=20, colors=(255,0,0))
img=draw_bounding_boxes(img, bbox1, width=20, colors=(255,0,0))
img=draw_bounding_boxes(img, bbox2, width=20, colors=(255,0,0))

# transform it to PIL image and display
img = torchvision.transforms.ToPILImage()(img)

#Label 

myFont = ImageFont.truetype('Arial.ttf',90)
I1 = ImageDraw.Draw(img)
I1.text(xy=(5600, 3200),text= "Road Sign", fill=(255, 0, 0),font=myFont,)
I1.text(xy=(2500, 4300),text= "Sealed Crack", fill=(255, 0, 0),font=myFont,)
I1.text(xy=(4200, 3900),text= "Delamination", fill=(255, 0, 0),font=myFont,)


img.show()
```
