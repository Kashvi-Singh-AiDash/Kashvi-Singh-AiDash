```python
from matplotlib import image
from matplotlib import pyplot as plt
import math
  
# to read the image stored in the working directory
data = image.imread('/Users/kashvi/Downloads/1907/Images/pano_0000_000088.jpg')


  
# to draw a line from (200,300) to (500,100)
x = [5682.5,5682.5 ]
y = [3191,4391]
plt.plot(x, y, color="red", linewidth=3)

#road (5000,4200)(6400,5000)
x1=[5000,6400]
y1= [4200,5000]
plt.plot(x1,y1,color="red",linewidth=3)

#Point where asset touches ground

yChange= int(y1[1]-y1[0])
xChange= int(x1[1]-x1[0])
slopeRoadLine = (yChange/xChange)

x2= x[0]
#y=mx+b
#b= y-mx
mx= int(x1[1]* slopeRoadLine)

b= (y1[1]- mx)
y2= int((slopeRoadLine * x[0]) + b)




# X distance from camera
x4 = [3200,x2]
y4 = [y2,y2]
plt.plot(x4, y4, color="blue", linewidth=3)

# Y distance from camera 

x3 = [3200,3200]
y3 = [y2,6400]
plt.plot(x3, y3, color="blue", linewidth=3)


#Hypotenuse to camera
x5 = [3200,x2 ]
y5 = [6400,y2]
plt.plot(x5, y5, color="green", linewidth=3)



xChange=int(3200-(6400-x2))
yChange=int(6400-y2)
pxDistance= int(math.sqrt((xChange**2)+(yChange**2)))
#At y = 2000, 1900 pixels ~3m ~0.0016m/pixel
EstDistance= int(pxDistance * 0.0016)


plt.plot(x2, y2, marker='v', color="white")
print('x-value: ' + str(x2) + '     y-value: ' + str(y2))
print("x shift: " + str(xChange) + "     y shift:  "+ str(yChange))
print("Pixel Distance: " + str(pxDistance) + "   Estimated Distance:  "+ str(EstDistance) +"m")

plt.imshow(data)
plt.show()
```
